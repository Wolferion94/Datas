```python
import pandas as pd
import pathlib
import numpy as np
```

# Cargue data

```python
#Gesti√≥n y cargue de data

CURRENT_DIR = pathlib.Path("..").resolve() #Lee la ruta actual del proyecto
DATA_DIR = CURRENT_DIR.joinpath("Proyecto_Estadistica","Data") #Establece la ruta de la data

file_base = list(DATA_DIR.glob("*.csv")) #Ruta del archivo enviado por Meli
```

```python
file_base[0]
```

```python
count_list = []
df_out = pd.DataFrame()

for file in file_base:

    file_ = str(file_base[0]).split("\\")[-1]
    file_ = file_.replace(".csv","")
    
    aux_df = pd.read_csv(file)
    aux_df['File'] = file_

    df_out = pd.concat([df_out.reset_index(drop=True),aux_df.reset_index(drop=True)], axis=0)

    aux = pd.read_csv(file).shape
    count_list.append(aux)
    
    print("Base " + str(file),"cantidad de registros y columnas",str(aux))
```

```python
df_modelado = df_out[['Code', 'Faction', 'Class', 'KB', 'D', 'HK', 'DD', 'HD','Honor','Rol', 'Win']].copy() #Se dejan solo las variables acordadas
```

```python
df_modelado['Win'] = df_modelado['Win'].fillna(0) #Se remplazan los nulos en las variables objetivo
```

```python
df_modelado.Win.value_counts() #Se realiza el conteo de las clases presentes en la variable objetivo
```

```python
df_modelado = df_modelado.drop_duplicates() #Se eliminan los registros duplicados
```
